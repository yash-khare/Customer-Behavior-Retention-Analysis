```python
import pandas as pd, numpy as np

mart = pd.read_csv("data/churn_mart.csv")

# Basic derived features
mart['tenure_days'] = (pd.to_datetime("2024-12-31") - pd.to_datetime(mart['signup_date'])).dt.days
mart['aov'] = np.where(mart['orders_cnt']>0, mart['revenue']/mart['orders_cnt'], 0)
mart['log_revenue'] = np.log1p(mart['revenue'])

# One-hot encode
X = mart[['recency_days','orders_cnt','log_revenue','tenure_days','aov','age','region','income_band']].copy()
X = pd.get_dummies(X, columns=['region','income_band'], drop_first=True)
y = mart['churn_label'].astype(int)

X.to_csv("data/X.csv", index=False)
y.to_csv("data/y.csv", index=False)
```
